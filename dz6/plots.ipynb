```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_optim = pd.read_csv("subset.csv")
df_optim = df_optim.sample(n=1200, random_state=42) #Берём случайные 1200 строк для адекватных графиков

plt.figure(figsize=(10, 6))
df_optim.sort_values('Mileage', inplace=True)
plt.plot(df_optim['Mileage'], df_optim['Price'])
plt.title("Линейный график для Price и Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.savefig("line_plot.png")
```

```python
plt.figure(figsize=(10, 6))
df_optim['Year'].hist(bins=20)
plt.title("Гистограмма Year")
plt.xlabel("Year")
plt.ylabel("Частота")
plt.savefig("hist_plot.png")
```

```python
plt.figure(figsize=(10, 6))
df_optim['Brand'].value_counts().plot(kind='bar')
plt.title("Столбчатая диаграмма для Brand")
plt.xlabel("Brand")
plt.ylabel("Количество")
plt.savefig("bar_plot.png")
```

```python
plt.figure(figsize=(8, 8))
df_optim['Model'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title("Круговая диаграмма для Model (топ 10)")
plt.ylabel('')
plt.savefig("pie_plot.png")
```

```python
plt.figure(figsize=(10, 8))
numeric_df = df_optim.select_dtypes(include=[float, int])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Корреляционная матрица")
plt.savefig("correlation_matrix.png")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_optim['Price'], df_optim['Mileage'], alpha=0.5)
plt.title("Диаграмма рассеяния Price vs Mileage")
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.savefig("scatter_plot.png")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_optim[['Price', 'Mileage']])
plt.title("Boxplot для Price и Mileage")
plt.savefig("box_plot.png")
```
